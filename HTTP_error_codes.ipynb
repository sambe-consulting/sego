```python
with open("HTTP response status codes - HTTP _ MDN.html","r") as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
from yapf.yapflib.yapf_api import FormatCode 
from yapf.yapflib.yapf_api import FormatFile
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
ids = ["InformationResponses",
       "SuccessfulResponses",
       "RedirectionMessages",
       "ClientErrorResponses",
       "ServerErrorResponses"]
dls = soup.find_all("dl",attrs={'itemprop':None,"class":None})
```

```python
final = {}
import textwrap

def pack_comments(stub,comment,message):
    prefix = "  "
    preferredWidth = 40
    wrapper = textwrap.TextWrapper(initial_indent=prefix, width=preferredWidth,
                               subsequent_indent=' '*(len(prefix)-10))
    return stub+"\n "+message+"\n"+wrapper.fill(comment)
        

def template(template,data):
    with open(template,'r') as f:
        template_content = f.read()
    for key in data:
        template_content =template_content.replace("{{"+key+"}}",data[key])
    return template_content

def analyse_dl(dl):
    results = []
    dt = dl.find_all("dt")
    dd = dl.find_all("dd")
    for i,j in zip(dt,dd):
        code = i.find("code").contents[0]
        comments = j.contents
        results.append({"code":code,"comments":comments})
    return results

for dl,title in zip(dls,ids):
    results = analyse_dl(dl)
    final[title] = results
```

```python
def build_classes(data,title):
    with open("sego/Exceptions/"+title+".py","w+") as f:
            f.write("from .SegoBaseException import SegoBaseException")
            f.write("\n")
    for code in data:
        class_stub,message = clean_code(code["code"])
        comment = clean_comments(code["comments"])
        comment = "\t\t "+comment 
        final = pack_comments(class_stub,comment,message)
        with open("sego/Exceptions/"+title+".py","a+") as f:
            f.write(final)
            f.write("\n\n\n")
        
    
def clean_code(code):
    split = code.split()
    message = "\tmessage='"+' '.join(split).replace("'","")+"'"
    name = ''.join(split[1:]).replace("-","").replace("'","")
    print(split[1])
    print(name)
    _class = "class "+name+"Exception"+str(split[0])+"(SegoBaseException):"
    return _class,message

def clean_comments(comments):
    cleaned = ""
    for com in comments:
        cleaned = cleaned+" "+str(com)

    cleaned = ' '.join(cleaned.split())
    cleaned = '""" '+cleaned+' """'
    return cleaned

    
    
    
# build_classes(final["InformationResponses"],"InformationResponses")
```

```python
for id in ids:
    build_classes(final[id],id)
```
